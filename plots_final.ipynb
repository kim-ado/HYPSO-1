### Component Substitution

```python
from numpy.linalg import svd
from skimage.exposure import match_histograms

image = example_cube.l1a_cube
sh_ind = example_cube.center_wavelength

img_variable_form = np.reshape(image, (image.shape[0] * image.shape[1], image.shape[2]))
U, S, Vh = svd(img_variable_form, full_matrices=False)
principal_components = np.dot(U, np.diag(S))
component_cube = np.reshape(principal_components, (image.shape[0], image.shape[1], image.shape[2]))
matched_sharpest_band = match_histograms(image[:,:,47], component_cube[:,:,0])
```

```python
# Load results from h1 
import pickle
results_file = "fwhm-lsf_h1_shalg.pkl"
with open(results_file, 'rb') as f:
    results = pickle.load(f)


valid_keys = [
    "component_substitution",
    "baseline",
    "laplacian_fo1",
    "wavelet_sym12_lvl2_regular",
    "wavelet_sym12_lvl3_regular",
    "wavelet_haar_lvl2_regular",
    "wavelet_haar_lvl3_regular",
    "wavelet_sym5_lvl2_regular",
    "wavelet_sym5_lvl3_regular",
    ]

Key_name_map = {
    "component_substitution": "PCA-CS",
    "baseline": "Baseline",
    "laplacian_fo1": "Laplacian",
    "wavelet_sym12_lvl2_regular": "Sym12-2",
    "wavelet_sym12_lvl3_regular": "Sym12-3",
    "wavelet_haar_lvl2_regular": "Haar-2",
    "wavelet_haar_lvl3_regular": "Haar-3",
    "wavelet_sym5_lvl2_regular": "Sym5-2",
    "wavelet_sym5_lvl3_regular": "Sym5-3",
    }


def make_boxplot_df(results, df_str):
    """Make a dataframe for boxplotting"""
    import pandas as pd
    df = pd.DataFrame()
    for key in results.keys():
        if key not in valid_keys:
            continue
        if df_str == "ssim_scores":
            res = results[key]["ssim_scores"]
            new_key = Key_name_map[key]
            df[new_key] = res
        elif "sam" in df_str:
            new_key = Key_name_map[key]
            df[new_key] = np.array(results[key][df_str])
        else:
            scores = results[key][df_str]
            out = np.mean(scores, axis=1)
            new_key = Key_name_map[key]
            df[new_key] = np.array(out)
    return df

# bp_df = make_boxplot_df(results,"fwhm_df")
bp_df = make_boxplot_df(results,"brisque_scores_df")

# Plot for brisque scores
import plotly.express as px
fig = px.violin(bp_df)
fig.update_layout(
    # title="Brisque Scores",
    xaxis_title="Sharpness Enhancement Algorithm",
    yaxis_title="Brisque Score [0-100]",
)
#
# make thight
fig.update_layout(
    autosize=False,
    width=800,
    height=300,
    margin=dict(
        l=50,
        r=50,
        b=10,
        t=10,
        pad=4
    ),
)
fig.show()
# save image
fig.write_image("fig/h1_brisque_scores.pdf")

# Plot for fwhm
bp_df = make_boxplot_df(results,"fwhm_df")

fig = px.violin(bp_df)
fig.update_layout(
    # title="FWHM",
    xaxis_title="Sharpness Enhancement Algorithm",
    yaxis_title="Estimated FWHM",
)
#
# make thight
fig.update_layout(
    autosize=False,
    width=800,
    height=300,
    margin=dict(
        l=50,
        r=50,
        b=10,
        t=10,
        pad=4
    ),
)
fig.show()
# save image
fig.write_image("fig/h1_fwhm.pdf")

# Plot for ssim

# bp_df = make_boxplot_df(results,"ssim_scores")

# fig = px.violin(bp_df)
# fig.update_layout(
#     # title="SSIM Scores",
#     xaxis_title="Sharpness Enhancement Algorithm",
#     yaxis_title="SSIM Score [0-1]",
# )
# #
# # make thight
# fig.update_layout(
#     autosize=False,
#     width=800,
#     height=300,
#     margin=dict(
#         l=50,
#         r=50,
#         b=10,
#         t=10,
#         pad=4
#     ),
# )
# fig.show()
# # save image
# fig.write_image("fig/h1_ssim_scores.pdf")
```
